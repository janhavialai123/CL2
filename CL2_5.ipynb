# Implement Agglomerative hierarchical clustering algorithm using appropriate dataset.

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
data=load_iris()
X=data.data
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
n_clusters=3
agg_clustering=AgglomerativeClustering(n_clusters=n_clusters,affinity='euclidean',linkage='ward')
labels=agg_clustering.fit_predict(X_scaled)
print("Cluster labels: ",labels)

```

```python
linked=linkage(X_scaled,method='ward')
plt.figure(figsize=(10,6))
dendrogram(linked,orientation='top',labels=data.target,distance_sort='descending',show_leaf_counts=True)
plt.title("Dendrogram for Agglomerative Hierarchical Clustering")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```
